```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
```

```python
#Checking working directory
os.getcwd()
```

```python
#Loading the csv file onto a pandas dataframe
df = pd.read_csv("dengue_features_train.csv")
```

```python
#Displaying the top 3 rows of data for better understanding
df.head(3)
```

```python
df.info()
```

```python
#Creating a copy of the original data-frame to avoid making changes to the original
df_rn = df.copy()
```

```python
#Calling the str.replace function to replace all column names containing 'station' with 'stn'
df_rn.columns = df_rn.columns.str.replace('station', 'stn')
```

```python
#Calling the head to look at the data and verify if the string replacement has been done properly
df_rn.head(3)
```

```python
df_rn.columns = df_rn.columns.str.replace('precip', 'prec')
df_rn.columns = df_rn.columns.str.replace('humidity', 'hd')
df_rn.columns = df_rn.columns.str.replace('reanalysis', 're_an')

#Listing out column names to check if the replacements have been effected. Not using head() as some column names are hidden from view
#Using the info() function to combine the answers for questions 1)e) and 1)f)
df_rn.info()
```

```python
#Converting the column 'Year' to category and verifying the same~
df_rn.year = df_rn.year.astype('category')
df_rn.dtypes
```

```python
df_pred = pd.read_csv("dengue_labels_train.csv")
df_pred.head(3)
```

```python
df_merged = pd.merge(df_rn, df_pred, on=['city','year','weekofyear'], how='outer')
```

```python
df_merged
```

```python
df_merged.info()
```

```python
#Counting the no. of NaN values in the dataframe
df_merged.isna().sum()
```

```python
#Using forward fill to fill the NaNs 
df_merged = df_merged.fillna(method='ffill')
```

```python
#Checking to see if all NaNs have been filled
df_merged.isna().sum()
```

```python
#Checking if all the values are same even after merging
df_pred[:3]
```

```python
df_rn[:3]
```

```python
df_merged[:3]
```

```python
#Scatterplot of 'weekofyear' vs. 'totalcases'
scat1 = df_merged.plot.scatter(x='weekofyear',
                              y='total_cases',
                              title='Relationship between Total Cases & Weeks of the Year')
scat1.set_xlabel("Weeks of the Year")
scat1.set_ylabel("Total Cases")
plt.show()

# From the scatterplot seen below, it looks like there is a weak positive relationship between the total cases and 
# weeks of the year with. The count of cases increases for certain years between the 30th and 50th weeks
```

```python
df_boxp = df_merged.sample(n=110, random_state=786)

box1 = df_boxp.boxplot(column='total_cases', by='city', vert=False, figsize=(15,5))
box1.set_xlabel("Total Cases")
box1.set_ylabel("City")
plt.title("Total Cases grouped by City for a sample of 110 records")
plt.suptitle("")

box2 = df_merged.boxplot(column='total_cases', by='city', vert=False, figsize=(15,5))
box2.set_xlabel("Total Cases")
box1.set_ylabel("City")
plt.title("Total Cases grouped by City for entire dataframe")
plt.suptitle("") 

# As we can see below, for a large enough sample the mean of the sample seems to represent the mean of the 
# population.
```

```python
df_grouped = df_merged.groupby('city')['total_cases']
#Mean of total_cases by city
print('Mean of total cases by city', '\n', df_grouped.mean(),'\n')

#Min of total_cases by city
print('Minimum of total cases by city', '\n', df_grouped.min(), '\n')
print('Maximum of total cases by city', '\n', df_grouped.max(), '\n')
print('Standard deviation of total cases by city', '\n', df_grouped.std(), '\n')
```

```python
hist1 = df_merged.hist(column = 'total_cases', by = 'city', figsize=(15,5), bins=10)
plt.xlabel('Total Cases')
plt.ylabel('Frequency')
plt.title('Histogram of Total Cases by City')

# As we can see here, the majority of the data is within two standard deviations of the mean and the rest of the 
# data is contained in one of the tails
```

```python
df_iq = df_merged[df_merged['city'] == 'iq']
```

```python
df_iq.pivot_table(index = ["year"], values = ["total_cases"], aggfunc=np.sum)

# There are 9 years with a total of more than or equal to 50 Dengue cases. The years are 2002, 2003, 2004, 2005,
# 2006, 2007, 2008, 2009 and 2010
```

```python
df_temp = df_merged.groupby('weekofyear')['total_cases']
df_temp
```

```python
df_temp.plot(kind = 'hist', legend = True, figsize=(20,20))
plt.xlabel('Total Cases by Week')
plt.ylabel('Frequency')
plt.title('Histogram of Total Cases by Weeks of the Year')
plt.show()

# Based on the histogram, I can see that the data for most of the weeks is skewed to the right
```
